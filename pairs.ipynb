-.diff() calculates the difference between consecutive log prices. This computes the log returns for each day compared to the previous day. The result is

-function is used to test for cointegration between two time series (s1 and s2) using the CADF test. It calculates the p-value for both possible orders of the time series and returns the smallest p-value, which is often used to determine whether the two time series are cointegrated.

-Cointegration allows for the construction of hedging strategies. If you have a portfolio of assets and you expect them to move together in the long run, you can use cointegrated assets to hedge against specific risks. 

-Cointegrated time series can provide more stable and accurate forecasts because they capture the long-term relationships and trends between variables, making it easier to predict future movements.

-Cointegration is widely used in pairs trading strategies, where traders look for pairs of assets (stocks, commodities, etc.) that are cointegrated. Cointegrated pairs exhibit a long-term statistical relationship, which means that the spread between their prices tends to revert to a stable mean. This can be exploited by buying the underperforming asset and shorting the outperforming one, profiting from the spread's mean-reverting behavior.

-In pair trading, you would typically use the CADF test to assess whether the price movements of two assets are cointegrated. If the CADF test suggests cointegration, it implies that the assets move together in the long run, which is a fundamental requirement for pair trading strategies



calculate_halflife:

-this function uses OLS(ordinary least square) regression to estimate the speed of mean reversion of a spread time series, and it returns the half-life, which indicates how quickly the spread reverts to its mean value after a deviation. This concept is often used in statistical arbitrage and pairs trading strategies to determine how long it takes for a spread to return to its equilibrium level.

-Pairs Selection: When constructing a portfolio of trading pairs, calculating the half-life helps in selecting pairs that exhibit a desirable mean-reversion behavior. Pairs with a shorter half-life may be chosen for shorter-term trading strategies, while longer half-lives are suitable for longer-term investments.

-Pairs Trading: The half-life concept is fundamental in pairs trading, a popular strategy in which traders identify pairs of assets with cointegrated spreads. The half-life determines the trading horizon and the timing of entry and exit points in such strategies.




ADF P-Value (Augmented Dickey-Fuller Test P-Value): The Augmented Dickey-Fuller test is a statistical test used to determine whether a time series is stationary (has a constant mean and variance). The ADF P-value is one of the outcomes of this test. It assesses the null hypothesis that a unit root is present in the time series data. A low ADF P-value (typically less than 0.05) indicates that the time series is likely stationary, which is a desirable property for many financial time series.

Spread Standard Deviation (Spread SD): A higher spread standard deviation suggests higher price volatility and may be of interest to traders who want to measure risk.

-This metric measures the standard deviation of the spread between two assets. The spread is the price difference between the two assets, and the standard deviation quantifies the variation or volatility in this spread. A higher spread standard deviation suggests higher price volatility and may be of interest to traders who want to measure risk.

Number of Zero-Crossings (Num Zero-Crossings): 
In the context of financial time series, this metric counts the number of times the spread or another relevant variable crosses zero. It can be a simple way to identify turning points or reversals in the spread's direction. For example, a spread that frequently crosses zero might be indicative of mean-reverting behavior.

Hurst Exponent: 
The Hurst exponent is a measure of a time series' long-term memory or self-similarity. It provides insights into whether a time series is trending (Hurst exponent > 0.5), mean-reverting (Hurst exponent < 0.5), or follows a random walk (Hurst exponent = 0.5). It is often used to characterize the behavior of time series data.

Half-Life of Mean Reversion:
The half-life of mean reversion measures how quickly a time series or spread reverts to its mean or equilibrium level after a deviation. It is often used in pairs trading and statistical arbitrage strategies to determine the trading horizon and the timing of entry and exit points. A shorter half-life suggests faster mean reversion.

Percentage of Days Within 2-SD Band (% Days Within 2-Standard Deviation Band): 
This metric calculates the percentage of days when the spread (or other relevant variable) remains within a 2-standard deviation (2-SD) band around its mean. A higher percentage indicates that the spread is within a more narrow range, which can be valuable for risk management and identifying trading opportunities.



```python
def calculate_distances(cumret):
    '''
    calculate Euclidean distance for each pair of stocks in the dataframe
    return sorted dictionary (in ascending order)
    '''
    distances = {} # dictionary with distance for each pair
    
    # calculate distances
    for s1 in cumret.columns:
        for s2 in cumret.columns:
            if s1!=s2 and (f'{s1}-{s2}' not in distances.keys()) and (f'{s2}-{s1}' not in distances.keys()):
                dist = np.sqrt(np.sum((cumret[s1] - cumret[s2])**2)) # Euclidean distance
                distances[f'{s1}-{s2}'] = dist
    
    # sort dictionary
    sorted_distances = {k:v for k,v in sorted(distances.items(), key = lambda item: item[1])}
    
    return sorted_distances

def parse_pair(pair):
    '''
    parse pair string S1-S2
    return tickers S1, S2
    '''
    dp = pair.find('-')
    s1 = pair[:dp]
    s2 = pair[dp+1:]
    
    return s1,s2


def cadf_pvalue(s1, s2, cumret):
    '''
    perform CADF cointegration tests
    since it is sensitive to the order of stocks in the pair, perform both tests (s1-2 and s2-s1)
    return the smallest p-value of two tests
    '''
    from statsmodels.tsa.stattools import coint
    
    p1 = coint(cumret[s1], cumret[s2])[1]
    p2 = coint(cumret[s2], cumret[s1])[1]
    
    return min(p1,p2)

def calculate_halflife(spread):
    '''
    calculate half-life of mean reversion of the spread
    '''
    from statsmodels.regression.linear_model import OLS
    from statsmodels.tools.tools import add_constant
    
    ylag = spread.shift()
    deltay = spread - ylag
    ylag.dropna(inplace=True)
    deltay.dropna(inplace=True)

    res = OLS(deltay, add_constant(ylag)).fit()
    halflife = -np.log(2)/res.params[0]
    
    return halflife

def calculate_metrics(sorted_distances, cumret, N=10):
    '''
    calculate metrics for N pairs with the smallest Euclidean distance
    return dataframe of results
    '''
    from hurst import compute_Hc
    from statsmodels.tsa.stattools import adfuller
    
    pairs = [k for k,v in sorted_distances.items()][:N]
    
    cols = ['Euclidean distance','Correlation', 'CADF p-value', 'ADF p-value', 'Spread SD', 
        'Num zero-crossings', 'Hurst Exponent', 'Half-life of mean reversion', '% days within 2-SD band']
    results = pd.DataFrame(index=pairs, columns=cols)
    
    for pair in pairs:
        s1,s2 = parse_pair(pair)
        spread = cumret[s1] - cumret[s2]
        results.loc[pair]['Euclidean distance'] = np.sqrt(np.sum((spread)**2))
        results.loc[pair]['Correlation'] = np.corrcoef(cumret[s1] , cumret[s2])[0, 1]
        results.loc[pair]['CADF p-value'] = cadf_pvalue(s1, s2, cumret)
        results.loc[pair]['ADF p-value'] = adfuller(spread)[1]
        results.loc[pair]['Spread SD'] = spread.std()
        results.loc[pair]['Num zero-crossings'] = ((spread[1:].values * spread[:-1].values) < 0).sum()
        results.loc[pair]['Hurst Exponent'] = compute_Hc(spread)[0]
        results.loc[pair]['Half-life of mean reversion'] = calculate_halflife(spread)
        results.loc[pair]['% days within 2-SD band'] = (abs(spread) < 2*spread.std()).sum() / len(spread) * 100
        
    return results

def plot_pairs(sorted_distances, cumret_train, cumret_test, N=5):
    '''
    plot cumulative returns of the spread for each of N pairs with smallest Euclidean distance
    '''
    pairs = [k for k,v in sorted_distances.items()][:N]
   
    for pair in pairs:
        s1,s2 = parse_pair(pair)
        spread_train = cumret_train[s1] - cumret_train[s2]
        spread_test = cumret_test[s1] - cumret_test[s2]
        spread_mean = spread_train.mean() # historical mean
        spread_std = spread_train.std() # historical standard deviation
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18,4))
        fig.suptitle(f'Spread of {pair} pair', fontsize=16)
        ax1.plot(spread_train, label='spread')
        ax1.set_title('Formation period')
        ax1.axhline(y=spread_mean, color='g', linestyle='dotted', label='mean')
        ax1.axhline(y=spread_mean+2*spread_std, color='r', linestyle='dotted', label='2-SD band')
        ax1.axhline(y=spread_mean-2*spread_std, color='r', linestyle='dotted')
        ax1.legend()
        ax2.plot(spread_test, label='spread')
        ax2.set_title('Trading period')
        ax2.axhline(y=spread_mean, color='g', linestyle='dotted', label='mean')
        ax2.axhline(y=spread_mean+2*spread_std, color='r', linestyle='dotted', label='2-SD band')
        ax2.axhline(y=spread_mean-2*spread_std, color='r', linestyle='dotted')
        ax2.legend()
```

```python
def select_pairs(train):
    '''
    select pairs using data from train dataframe
    return dataframe of selected pairs
    '''
    tested = []

    from statsmodels.regression.linear_model import OLS
    from statsmodels.tools.tools import add_constant
    from hurst import compute_Hc
    from statsmodels.tsa.stattools import adfuller

    cols = ['Euclidean distance', 'Num zero-crossings', 'Pearson r', 
            'Spread SD', 'Hurst Exponent', 'Half-life of mean reversion']
    pairs = pd.DataFrame(columns=cols)

    for s1 in train.columns:
        for s2 in train.columns:
            if s1!=s2 and (f'{s1}-{s2}' not in tested) and (f'{s2}-{s1}' not in tested):
                tested.append(f'{s1}-{s2}')
                cadf_p = cadf_pvalue(s1,s2,train)
                if cadf_p<0.05:
                    spread = train[s1] - train[s2]
                    hurst = compute_Hc(spread)[0]
                    if hurst<0.5:
                        halflife = calculate_halflife(spread)
                        if halflife>1 and halflife<30:
                            num_crossings = (spread.values[1:] * spread.values[:-1] < 0).sum()
                            if num_crossings>len(train.index)/252*12: 
                                distance = np.sqrt(np.sum((train[s1] - train[s2])**2)) # Euclidean distance
                                pearson_r = np.corrcoef(train[s1], train[s2])[0][1]
                                spread_sd = spread.std()
                                pairs.loc[f'{s1}-{s2}'] = [distance, num_crossings, pearson_r,
                                                           spread_sd, hurst, halflife]
                                
    return pairs
```

```python
def parse_pair(pair):
    '''
    parse pair string S1-S2
    return tickers S1, S2
    '''
    dp = pair.find('-')
    s1 = pair[:dp]
    s2 = pair[dp+1:]
    
    return s1,s2

def cadf_pvalue(s1, s2, cumret):
    '''
    perform CADF cointegration tests
    since it is sensitive to the order of stocks in the pair, perform both tests (s1-2 and s2-s1)
    return the smallest p-value of two tests
    '''
    from statsmodels.tsa.stattools import coint
    
    p1 = coint(cumret[s1], cumret[s2])[1]
    p2 = coint(cumret[s2], cumret[s1])[1]
    
    return min(p1,p2)

def calculate_halflife(spread):
    '''
    calculate half-life of mean reversion of the spread
    '''
    from statsmodels.regression.linear_model import OLS
    from statsmodels.tools.tools import add_constant
    
    ylag = spread.shift()
    deltay = spread - ylag
    ylag.dropna(inplace=True)
    deltay.dropna(inplace=True)

    res = OLS(deltay, add_constant(ylag)).fit()
    halflife = -np.log(2)/res.params[0]
    
    return halflife

def calculate_metrics(pairs, cumret, pairs_df):
    '''
    calculate metrics for pairs using data in cumret
    return dataframe of results
    '''
    from hurst import compute_Hc
    from statsmodels.tsa.stattools import adfuller
    
    cols = ['Euclidean distance', 'CADF p-value', 'ADF p-value', 'Spread SD', 'Pearson r',
        'Num zero-crossings', 'Hurst Exponent', 'Half-life of mean reversion', '% days within historical 2-SD band']
    results = pd.DataFrame(index=pairs, columns=cols)
    
    for pair in pairs:
        s1,s2 = parse_pair(pair)
        spread = cumret[s1] - cumret[s2]
        results.loc[pair]['Euclidean distance'] = np.sqrt(np.sum((spread)**2))
        results.loc[pair]['CADF p-value'] = cadf_pvalue(s1, s2, cumret)
        results.loc[pair]['ADF p-value'] = adfuller(spread)[1]
        hist_sd = pairs_df.loc[pair]['Spread SD'] # historical standard deviation
        results.loc[pair]['Spread SD'] = hist_sd
        results.loc[pair]['Pearson r'] = np.corrcoef(cumret[s1], cumret[s2])[0][1]
        results.loc[pair]['Num zero-crossings'] = ((spread[1:].values * spread[:-1].values) < 0).sum()
        results.loc[pair]['Hurst Exponent'] = compute_Hc(spread)[0]
        results.loc[pair]['Half-life of mean reversion'] = calculate_halflife(spread)
        results.loc[pair]['% days within historical 2-SD band'] = (abs(spread) < 2*hist_sd).sum() / len(spread) * 100
        
    return results

def plot_pairs(pairs, cumret_train, cumret_test):
    '''
    plot cumulative returns of the spread for each pair in pairs
    '''
    
    for pair in pairs:
        s1,s2 = parse_pair(pair)
        spread_train = cumret_train[s1] - cumret_train[s2]
        spread_test = cumret_test[s1] - cumret_test[s2]
        spread_mean = spread_train.mean() # historical mean
        spread_std = spread_train.std() # historical standard deviation

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18,4))
        fig.suptitle(f'Spread of {pair} pair', fontsize=16)
        ax1.plot(spread_train, label='spread')
        ax1.set_title('Formation period')
        ax1.axhline(y=spread_mean, color='g', linestyle='dotted', label='mean')
        ax1.axhline(y=spread_mean+2*spread_std, color='r', linestyle='dotted', label='2-SD band')
        ax1.axhline(y=spread_mean-2*spread_std, color='r', linestyle='dotted')
        ax1.legend()
        ax2.plot(spread_test, label='spread')
        ax2.set_title('Trading period')
        ax2.axhline(y=spread_mean, color='g', linestyle='dotted', label='mean')
        ax2.axhline(y=spread_mean+2*spread_std, color='r', linestyle='dotted', label='2-SD band')
        ax2.axhline(y=spread_mean-2*spread_std, color='r', linestyle='dotted')
        ax2.legend()

top5_distance = list(pairs12_6.sort_values(by='Euclidean distance').index[:5])
```

```python

def pnl(pairs,data):
    short_window = 5
    long_window = 60

    for pair_name, pair_info in pairs.iterrows():
        s1, s2 = pair_name.split('-')
        S1 = data[s1]
        S2 = data[s2]
        ratios = S1/S2
        len(ratios)
        train12_6 = ratios[:300] # formation period
        ratios_mavg5 =train12_6.rolling(window=short_window, center=False).mean()
        ratios_mavg60 =train12_6.rolling(window=long_window, center=False).mean()
        std_60 =train12_6.rolling(window=long_window, center=False).std()
        
        # Calculate the Z-score for the pair
        zscore=  (ratios_mavg5 - ratios_mavg60) / std_60
        buy =train12_6.copy()
        sell =train12_6.copy()
        buy[zscore>-1] = 0
        # Plot the prices and buy and sell signals from z score
        plt.figure(figsize=(18,9))
        S1[60:].plot(color='b')
        S2[60:].plot(color='c')
        buyR = 0*S1.copy()
        sellR = 0*S1.copy()
        # When buyving the ratio, buy S1 and sell S2
        buyR[buy!=0] = S1[buy!=0]
        sellR[buy!=0] = S2[buy!=0]
        # When selling the ratio, sell S1 and buy S2 
        buyR[sell!=0] = S2[sell!=0]
        sellR[sell!=0] = S1[sell!=0]
        buyR[60:].plot(color='g', linestyle='None', marker='^')
        sellR[60:].plot(color='r', linestyle='None', marker='v')
        x1,x2,y1,y2 = plt.axis()
        plt.axis((x1,x2,min(S1.min(),S2.min()),max(S1.max(),S2.max())))
        plt.legend([f'{s1}',f'{s2}', 'Buy Signal', 'Sell Signal'])
        plt.show()
        money = 0
        countS1 = 0
        countS2 = 0
        for i in range(len(ratios)):
            # Sell short if the z-score is > 1
                if zscore[i] > 1:
                    money += S1[i] - S2[i] * ratios[i]
                    countS1 -= 1
                    countS2 += ratios[i]
                    # print('Selling Ratio %s %s %s %s'%(money, ratios[i], countS1,countS2))
                # Buy long if the z-score is < 1
                elif zscore[i] < -1:
                    money -= S1[i] - S2[i] * ratios[i]
                    countS1 += 1
                    countS2 -= ratios[i]
                    # print('Buying Ratio %s %s %s %s'%(money,ratios[i], countS1,countS2))
                # Clear positions if the z-score between -.5 and .5
                elif abs(zscore[i]) < 0.75:
                    money += S1[i] * countS1 + S2[i] * countS2
                    countS1 = 0
                    countS2 = 0
                    # print('Exit pos %s %s %s %s'%(money,ratios[i], countS1,countS2))
                    
          
```

```python
def trade(pairs,data):
    window1 = 5 
    window2 = 60
    for pair_name, pair_info in pairs.iterrows():
        s1, s2 = pair_name.split('-')
        S1 = data[s1].iloc[:300]
        S2 = data[s2].iloc[:300]
        # If window length is 0, algorithm doesn't make sense, so exit
        if (window1 == 0) or (window2 == 0):
            return 0
        
        # Compute rolling mean and rolling standard deviation
        ratios = S1/S2
        ma1 = ratios.rolling(window=window1,
                                center=False).mean()
        ma2 = ratios.rolling(window=window2,
                                center=False).mean()
        std = ratios.rolling(window=window2,
                            center=False).std()
        zscore = (ma1 - ma2)/std
        
        # Simulate trading
        # Start with no money and no positions
        money = 0
        countS1 = 0
        countS2 = 0
        for i in range(len(ratios)):
            # Sell short if the z-score is > 1
            if zscore[i] > 1:
                money += S1[i] - S2[i] * ratios[i]
                countS1 -= 1
                countS2 += ratios[i]
                # print('Selling Ratio %s %s %s %s'%(money, ratios[i], countS1,countS2))
            # Buy long if the z-score is < 1
            elif zscore[i] < -1:
                money -= S1[i] - S2[i] * ratios[i]
                countS1 += 1
                countS2 -= ratios[i]
                # print('Buying Ratio %s %s %s %s'%(money,ratios[i], countS1,countS2))
            # Clear positions if the z-score between -.5 and .5
            elif abs(zscore[i]) < 0.75:
                money += S1[i] * countS1 + S2[i] * countS2
                countS1 = 0
                countS2 = 0
                # print('Exit pos %s %s %s %s'%(money,ratios[i], countS1,countS2))
        return money*100                
```

```python
trade(top5_pairs_df,train12_6)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

def linear_regression(pairs,cumret):
    for pair_name, pair_info in pairs.iterrows():
        s1, s2 = pair_name.split('-')

        X = cumret[s1].values.reshape(-1, 1)
        y = cumret[s2]

# Create and fit the linear regression model
        model = LinearRegression()
        model.fit(X, y)

        # Predict the target stock prices
        y_pred = model.predict(X)

        # Plot the actual and predicted stock prices
        plt.figure(figsize=(12, 6))
        plt.scatter(cumret.index, y, label='Actual ' + s2 + ' Price', color='blue')
        plt.plot(cumret.index, y_pred, label='Predicted ' + s2 + ' Price', color='red')
        plt.title('Linear Regression for Stock Price Prediction')
        plt.xlabel('Date')
        plt.ylabel('Stock Price')
        plt.legend()
        plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pmdarima import auto_arima

def arima_model(pairs, cumret):
    for pair_name, pair_info in pairs.iterrows():
        s1, s2 = pair_name.split('-')

        # Step 1: Prepare the data
        ratio = cumret[s1] / cumret[s2]

        # Step 2: Train the ARIMA model
        model = auto_arima(ratio, seasonal=False, stepwise=True, trace=True)

        # Step 3: Make predictions
        forecast_steps = 50  # Change this to the desired number of forecast steps
        forecast, conf_int = model.predict(n_periods=forecast_steps, return_conf_int=True)

        # Step 4: Plot the results
        plt.figure(figsize=(12, 6))
        plt.plot(cumret.index, ratio, label='Actual Ratio', color='blue')
        plt.plot(range(len(ratio), len(ratio) + forecast_steps), forecast, label='Forecasted Ratio', color='red')
        plt.fill_between(range(len(ratio), len(ratio) + forecast_steps), conf_int[:, 0], conf_int[:, 1], color='pink', alpha=0.5, label='95% Confidence Interval')
        plt.title(f'ARIMA Model for Stock Price Prediction {pair_name}')
        plt.xlabel('Date')
        plt.ylabel('Stock Ratio')
        
        # Corrected code to set x-axis ticks
        x_ticks = list(range(len(ratio))) + list(range(len(ratio), len(ratio) + forecast_steps))
        last_date = cumret.index[-1]
        x_labels = cumret.index.tolist() + [last_date + pd.DateOffset(days=i) for i in range(1, forecast_steps + 1)]
        x_labels = [str(date) for date in x_labels]  # Convert dates to strings
        plt.xticks(x_ticks, x_labels, rotation=45)  # Rotate x-axis labels for better visibility

        plt.legend()
        plt.show()

# Example usage
# Assuming you have the 'pairs' DataFrame from your dataset
# and 'cumret' DataFrame for cumulative returns
arima_model(top5_pairs_df, cumret)
```
